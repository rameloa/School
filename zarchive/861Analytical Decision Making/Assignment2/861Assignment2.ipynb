```python
import numpy as np
import pandas as pd

# Function to simulate a single iteration
def simulate_day(num_csts):
    # Generate the number of hardware problems
    num_problems = np.random.binomial(num_computers, prob_problem)
    
    # Generate the number of major problems
    num_major_problems = np.random.binomial(num_problems, prob_major_problem)
    
    # Calculate the total time required
    total_time_needed = num_problems * time_minor_problem + num_major_problems * time_major_problem
    
    # Calculate the total available time
    total_available_time = num_csts * work_hours_per_cst
    
    # Calculate the cost
    if total_time_needed > total_available_time:
        external_support_hours = total_time_needed - total_available_time
        external_support_cost_total = external_support_hours * external_support_cost
    else:
        external_support_hours = 0
        external_support_cost_total = 0
    
    total_cost = num_csts * cost_per_cst + external_support_cost_total
    return total_cost, external_support_hours
```

```python
# Question 1

# Parameters
num_computers = 200
prob_problem = 0.05
time_minor_problem = 0.75
prob_major_problem = 0.30
time_major_problem = 2
work_hours_per_cst = 6
cost_per_cst = 300
external_support_cost = 95
sample_size = 1000

# Simulate for different numbers of CSTs
results = []
for num_csts in range(1, 5):
    total_costs = []
    total_external_support_hours = []
    
    for _ in range(sample_size):
        cost, external_support_hours = simulate_day(num_csts)
        total_costs.append(cost)
        total_external_support_hours.append(external_support_hours)
    
    avg_cost = np.mean(total_costs)
    avg_external_support_hours = np.mean(total_external_support_hours)
    results.append((num_csts, avg_cost, avg_external_support_hours))

# Convert results to DataFrame
results_df = pd.DataFrame(results, columns=['Number of CSTs', 'Average Cost', 'Average External Support Hours'])

# Display the results
print(results_df)
```
